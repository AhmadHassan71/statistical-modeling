# Workshop 8: Polynomial Regression On Synthetic Data

## Importing Libraries

```python
import numpy as np
import pymc3 as pm
import matplotlib.pyplot as plt
```

### Generating Data

```python
np.random.seed(42)  # For consistency and reproducibility
                    # to ensure you get same random numbers

n_samples=100                  # number of samples
X=np.linspace(0,10,n_samples)  # creates equally spaced samples 
                               # from 0 to 10

# creates a quadratic dataset with some random noise
y= 3*X**2 +2*X + 5 + np.random.normal(scale=10,size=n_samples)
y_cubic= 3*X**3+2*X**2+5*X+10+ np.random.normal(scale=10,size=n_samples)

# adding noise adds some realism allowing to test data for
# real world problems 
```

### Defining the model

**Quadratic Model**

```python
degree = 2

with pm.Model() as model:

    # There are degree + 1 coeffs
    coeffs= pm.Normal('coeffs', mu=0, sd=10, shape=degree+1) 

    # This calculates the dot product between the coeffs array
    # and the transpoed Vandermonde matrix
    # the vandermode matrix is a matrix that contains the
    # different powers of the input variable
    # the increasing= true ensures that this increases from
    # left to right
    y_pred=pm.math.dot(coeffs,np.vander(X,degree+1, increasing=True).T)

    likelihood= pm.Normal('y_obs', mu= y_pred, sd=10, observed=y)

    trace= pm.sample(2000, tune=1000)
```

### Visualising the results

```python
pm.traceplot(trace)
plt.show()
```

**Plotting the fitted polynomial curve**

```python
x_vals= np.linspace(0,10,100)
# dot product of coeffics matrix and the transposed vandermonde
# x_vals matrix
y_fitted = np.dot(np.vander(x_vals,degree+1,increasing=True),trace['coeffs'].T)

plt.scatter(X,y,label='Data')
plt.plot(x_vals,y_fitted,color='r',alpha=0.1)
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

### Cubic Model

```python
degree = 3

with pm.Model() as model2:

    coeffs=pm.Normal('coeffs', mu=0,sd=10,shape=degree+1)
    y_pred= pm.math.dot(coeffs,np.vander(X,degree+1,increasing=True).T)
    likelihood=pm.Normal('y_obs',mu=y_pred,sd=10,observed=y_cubic)

    trace=pm.sample(2000,tune=2000)
```

```python
x_vals = np.linspace(0, 10, 100)  # New X values for prediction
y_fitted = np.dot(np.vander(x_vals, degree + 1, increasing=True), trace['coeffs'].T)

plt.scatter(X, y_cubic, label='Data')
plt.plot(x_vals, y_fitted, color='r', alpha=0.1)
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```
